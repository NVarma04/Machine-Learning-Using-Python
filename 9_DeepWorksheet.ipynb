```python

import numpy as np

r3,c3=map(int,input('Enter the size of batch in rows and columns : ').split())
x=np.zeros((r3,c3))
for i in range(r3):
  for j in range(c3):
    x[i][j]=input('Enter: ')
print(x)

n=int(input('Enter the number of w matrix taken: '))
r1,c1=map(int,input('Enter the size of each w matrix in rows and columns : ').split())
r2,c2=map(int,input('Enter the size of each bias matrix in rows and columns : ').split())
for p in range(1,n+1):

  def relu(a):
    return np.maximum(0,a)

  print('Elements of {0} w matrix: '.format(p))
  w=np.zeros((r1,c1))
  for i in range(r1):
    for j in range(c1):
      w[i][j]=input('Enter: ')
  print(w)

  print('Elements of {0} b matrix: '.format(p))
  b=np.zeros((r2,c2))
  for i in range(r2):
    for j in range(c2):
      b[i][j]=input('Enter: ')
  print(b)

  if p==1:
    z=np.dot(w,x)
    y=z+b
    print('Result: ')
    print(y)
    print('ReLU result: ')
    re_val=relu(y)
    print(re_val)
  else:
    z=np.dot(w,re_val)
    y=z+b
    print('Result: ')
    print(y)
    print('ReLU result: ')
    re_val=relu(y)
    print(re_val)




```
