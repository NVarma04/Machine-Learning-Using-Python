```python
import numpy as np

r1,c1=map(int,input('Enter the size of w matrix in rows and columns : ').split())
print('Elements of w matrix: ')
w=np.zeros((r1,c1))
for i in range(r1):
  for j in range(c1):
    w[i][j]=input('Enter: ')

print('Connections are present between :\n')
for i in range(r1):
  for j in range(c1):
    if w[i][j]!=0:
      print('X{0} and Y{1}'.format(j+1,i+1))

```
